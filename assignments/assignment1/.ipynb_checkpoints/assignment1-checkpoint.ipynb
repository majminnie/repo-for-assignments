# Task 1: A list of dictionaries

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# a) Filter out invalid transactions (price == 0 or quantity < 0)
valid = [t for t in transactions if t["price"] > 0 and t["quantity"] > 0]

# b) Compute total = price * quantity for each valid transaction
for t in valid:
    t["total"] = t["price"] * t["quantity"]

# c) Group by region and compute total revenue per region
revenue = {}
for t in valid:
    revenue[t["region"]] = revenue.get(t["region"], 0) + t["total"]

# d) Print the region with the highest revenue using an f-string
top_region = max(revenue, key=revenue.get)
print(f"The region with the highest revenue is {top_region} with €{revenue[top_region]}")
```

# Task 2: Creating a function

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(sales):
    total = sum(sales)
    average = round(total / len(sales), 2)
    best_day = sales.index(max(sales)) + 1
    worst_day = sales.index(min(sales)) + 1
    above_average_count = len([s for s in sales if s > average])
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }

result = summarise_sales(sales)
print(result)
```

# Task 3: Loops

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# a) List of students with grade 50 or above
passing = [s["name"] for s in students if s["grade"] >= 50]
print(passing)

# b) Print status for each student
for s in students:
    if s["grade"] >= 80:
        print(f"{s['name']}: Excellent")
    elif s["grade"] >= 50:
        print(f"{s['name']}: Pass")
    else:
        print(f"{s['name']}: Fail")

# c) Count students with attendance below 0.80
low_attendance = 0
for s in students:
    if s["attendance"] < 0.80:
        low_attendance += 1
print(low_attendance)

# d) Find highest grade using a loop (no max())
top_student = students[0]
for s in students:
    if s["grade"] > top_student["grade"]:
        top_student = s
print(f"{top_student['name']}: {top_student['grade']}")

# e) Find highest grade using max()
top = max(students, key=lambda s: s["grade"])
print(f"{top['name']}: {top['grade']}")
```

# Task 4: Class and methods

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    # a) Add or update an asset
    def add_asset(self, ticker, shares, price_per_share):
        self.holdings[ticker] = {"shares": shares, "price_per_share": price_per_share}

    # b) Total market value
    def total_value(self):
        return sum(v["shares"] * v["price_per_share"] for v in self.holdings.values())

    # c) Most valuable asset
    def most_valuable(self):
        return max(self.holdings, key=lambda t: self.holdings[t]["shares"] * self.holdings[t]["price_per_share"])

    # d) Readable summary
    def __str__(self):
        return f"Owner: {self.owner}, Assets: {len(self.holdings)}, Total Value: €{self.total_value()}"

# Create instance and demonstrate
p = Portfolio("Maj")
p.add_asset("AAPL", 10, 180)
p.add_asset("TSLA", 5, 250)
p.add_asset("GOOGL", 3, 140)

print(p.total_value())
print(p.most_valuable())
print(p)
```

# Task 5: Importing a module

```python
from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

dates = [datetime.strptime(order[0], "%Y-%m-%d").date() for order in orders]

earliest = min(dates)
latest = max(dates)
days_between = (latest - earliest).days

print(f"Earliest order: {earliest}")
print(f"Latest order: {latest}")
print(f"Days between earliest and latest order: {days_between}")
```

# Task 6: Array operations and broadcasting

```python
import numpy as np

revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]
])

# a) Total annual revenue per year (row sums)
annual_revenue = np.sum(revenue, axis=1)
print(f"Annual revenue: {annual_revenue}")

# b) Mean quarterly revenue across all years (column means)
quarterly_means = np.mean(revenue, axis=0)
print(f"Mean quarterly revenue: {quarterly_means}")

# c) Broadcasting - apply growth adjustment to 2024 row
growth = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * growth
print(f"Adjusted 2024 values: {adjusted_2024}")

# d) Boolean indexing - values exceeding 250
above_250 = revenue[revenue > 250]
print(f"Values exceeding 250: {above_250}")
```

# Task 7: Pandas

```python
import pandas as pd

data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}
df = pd.DataFrame(data)

# a) Add salary_per_year_exp column
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round()

# b) Sort by performance_score descending, print top 3
top3 = df.sort_values("performance_score", ascending=False).head(3)
print(top3)

# c) Finance employees with performance_score above 4.0
finance_high = df.loc[(df["department"] == "Finance") & (df["performance_score"] > 4.0)]
print(finance_high)

# d) Rename years_exp to experience_years
df = df.rename(columns={"years_exp": "experience_years"})

# e) First four employees, salary and performance_score columns
print(df.iloc[:4][["salary", "performance_score"]])
```

# Task8: More pandas

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

# a) Create DataFrame with date as index
df = pd.DataFrame({
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
}, index=days)
df.index.name = "date"

# b) Compute conversions and daily_revenue
df["conversions"] = (df["visitors"] * df["conversion_rate"]).round().astype(int)
df["daily_revenue"] = df["conversions"] * df["avg_order_value"]

print(df.head())
```
